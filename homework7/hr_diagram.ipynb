```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)
```

```python
print(data.shape)
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
print("Magnitude min:", np.min(magnitude))
print("Magnitude max:", np.max(magnitude))

print("Temperature min:", np.min(temperature))
print("Temperature max:", np.max(temperature))
```

```python
brightest_index = np.argmin(magnitude)
dimmest_index = np.argmax(magnitude)

print(brightest_index)
print(dimmest_index)
```

```python
print(np.min(temperature))
print(np.max(temperature))

coldest_index = np.argmin(temperature)
hottest_index = np.argmax(temperature)
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(log_temperature, log_magnitude, label="Stars")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.scatter(log_temperature[brightest_index], log_magnitude[brightest_index],
            color="black", marker="*", s=300, label="Brightest")

plt.scatter(log_temperature[dimmest_index], log_magnitude[dimmest_index],
            color="purple", marker="*", s=300, label="Dimmest")

plt.scatter(log_temperature[hottest_index], log_magnitude[hottest_index],
            color="green", marker="*", s=300, label="Hottest")

plt.scatter(log_temperature[coldest_index], log_magnitude[coldest_index],
            color="orange", marker="*", s=300, label="Coldest")
```
